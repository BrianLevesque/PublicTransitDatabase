```python
import pandas as pd

cta = pd.read_csv('CTA_-_Ridership_-_Daily_Boarding_Totals_20240612.csv')

mta = pd.read_csv("MTA_Monthly_Ridership___Traffic_Data__Beginning_January_2008.csv")

cta
```

```python
cta.info()
```

```python
#converting service_date to datetime objects

cta['service_date'] = pd.to_datetime(cta['service_date'], format='%m/%d/%Y')

#creating a month column
cta['Month-Year'] = cta['service_date'].dt.to_period('M').astype(str)

#creating a new dataframe with groupby month
cta_cleaned = cta.groupby('Month-Year').agg({
    'bus': 'sum',
    'rail_boardings': 'sum'
    }).reset_index()


#creating the city column and setting it to chicago
cta_cleaned['city'] = 'Chicago'
cta_cleaned
```

```python
#seperating the df into 2 dfs for bus info and metro info
cta_bus = cta_cleaned[["Month-Year","bus","city"]]
cta_metro = cta_cleaned[["Month-Year","rail_boardings","city"]]



#cleaning and renaming to column names and creating columns transportation type and ridership_count
cta_bus["month"] = cta_bus["Month-Year"]
cta_bus['transportation_type'] = 'bus'
cta_bus['ridership_count'] = cta_bus['bus']
cta_bus = cta_bus.drop(columns=["Month-Year","bus"])

cta_metro["month"] = cta_metro["Month-Year"]
cta_metro['transportation_type'] = 'metro'
cta_metro['ridership_count'] = cta_metro['rail_boardings']
cta_metro = cta_metro.drop(columns=["Month-Year","rail_boardings"])
```

```python
mta['Agency'].value_counts()
```

```python
#converting service_date to datetime objects

mta['Month'] = pd.to_datetime(mta['Month'], format='%m/%d/%Y')

#creating a year and month columns
mta["Month-Year"] = mta['Month'].dt.to_period('M').astype(str)


#renaming the columns rearranging columns to fit desired naming conventions
mta['month'] = mta["Month-Year"]
```

```python
#making a df with just bus and subway information then adding a column for city
mta_cleaned = mta[mta['Agency'].isin(['Bus',"Subway"])]



mta_cleaned['city'] = 'New York City'

mta_cleaned.info()
```

```python
#splitting up the bus and subway numbers into seperate dataframes for each city

mta_bus = mta_cleaned[mta_cleaned['Agency'].isin(['Bus'])]

mta_metro = mta_cleaned[mta_cleaned['Agency'].isin(['Subway'])]

mta_bus['transportation_type'] = 'bus'
mta_bus['ridership_count'] = mta_bus['Ridership']

mta_bus = mta_bus.drop(columns=['Month','Agency','Ridership','Month-Year'])
mta_bus
```

```python
mta_metro['transportation_type'] = 'metro'
mta_metro['ridership_count'] = mta_metro['Ridership']

mta_metro = mta_metro.drop(columns=['Month','Agency','Ridership','Month-Year'])
mta_metro
```

```python
#concatenting the metro dfs for both cities
metro_df = pd.concat([mta_metro, cta_metro], axis = 0,ignore_index=True)

metro_df
```

```python
#concatenting the bus dfs for both cities
bus_df = pd.concat([mta_bus, cta_bus], axis = 0,ignore_index=True)

bus_df
```

```python
#concatenting all df together because i realized that they were all suppose to be together

monthly_totals = pd.concat([bus_df,metro_df],axis=0, ignore_index=True)

monthly_totals
```
